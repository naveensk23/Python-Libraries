# sweetviz

Sweetviz is an open source Python library that generates beautiful, high-density visualizations to kickstart EDA (Exploratory Data Analysis) with a single line of code. Output is a fully self-contained HTML application.

The best way to install sweetviz------------------>  pip install sweetviz   

```python
import pandas as pd
```

```python
import sweetviz
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
report = sweetviz.analyze([train, "train"])
```

```python
report = sweetviz.analyze([train, "Train"], target_feat = 'SalePrice')
```

```python
report.show_html('Report.html')
```

```python
report1 = sweetviz.compare([train, "Train"],[test,"Test"], "SalePrice")
```

```python
report1.show_html('Report.html')
```
